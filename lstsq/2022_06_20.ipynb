```python
import numpy as np
```

```python
A = np.random.rand(24).reshape(6,4)
```

```python
## QR decomposition
def factorize(A):
    width = A.shape[1]

    Q = np.zeros_like(A)
    R = np.zeros((width,width))

    ## first step
    norm_q = np.linalg.norm(A[:,0])
    Q[:,0] = A[:,0] / norm_q
    R[0,0] = norm_q

    ## gram-schmidts algorithm
    for i in range(1,width):
        Q[:,i] = A[:,i]
        
        for j in range(i):
            Q[:,i] -= np.dot(Q[:,j], A[:,i]) * Q[:,j]
            R[j,i] = np.dot(Q[:,j],A[:,i])

        norm_q = np.linalg.norm(Q[:,i])
        Q[:,i] = Q[:,i] / norm_q
        R[i,i] = norm_q

        ## linearly dependence
        if norm_q == 0:
            print("it has linearly dependent columns")
            break

    return Q, R
```

```python
Q, R = factorize(A)
```

```python
print(f"Q: {Q}")
print(f"R: {R}")
print(f"QR: {np.dot(Q,R)}")
print(f"{np.allclose(A, np.dot(Q,R))}")
```

```python
A = np.random.rand(32).reshape(8,4)
A1 = A[:4,:]
A2 = A[4:,:]
Q1, R1 = factorize(A1)
Q2, R2 = factorize(A2)
```

```python
print(f"Q1: {Q1}\n")
print(f"Q2: {Q2}\n")
print(f"R1: {R1}\n")
print(f"R2: {R2}")
```
